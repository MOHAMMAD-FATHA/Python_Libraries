```python
"""
* @Author: Mohammad Fatha
* @Date: 2021-10-10 02:30
* @Last Modified by: Mohammad Fatha
* @Last Modified time: 2021-10-10 02:30
* @Title: :Program to draw a scatter plot for random 1000 x and y coordinates
"""

# x and y given as array_like objec
import plotly.express as px
import random
data = []
data1 = []
for i in range(1000):
    x = random.randint(0,1000)
    data.append(x)

for i in range(1000):
    x = random.randint(0,1000)
    data1.append(x)

fig = px.scatter(x=data, y=data1)

fig.show()
```
