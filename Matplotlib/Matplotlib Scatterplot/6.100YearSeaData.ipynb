```python
"""
* @Author: Mohammad Fatha
* @Date: 2021-10-08 02:30
* @Last Modified by: Mohammad Fatha
* @Last Modified time: 2021-10-08 02:30
* @Title: : Python program to draw a scatter plot to find sea level rise in past 100 years. We have 
collected the data from here [https://github.com/datasets/sea-level-rise].
"""

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('data.csv')
data = pd.DataFrame(data)
year = data['year']
sea_levels = data['CSIRO_sea_level']
plt.scatter(year, sea_levels, c='red', edgecolors='g')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sealevel')
plt.show()
```
