```python
"""
* @Author: Mohammad Fatha
* @Date: 2021-10-07 00:09
* @Last Modified by: Mohammad Fatha
* @Last Modified time: 2021-10-07 00:09
* @Title: : Python program to plot two or more lines on same plot with suitable legends of each line.
"""

import matplotlib.pyplot as plt
# line 1 points
x1 = [10,20,30]
y1 = [20,40,10]
# line 2 points
x2 = [10,20,30]
y2 = [40,10,30]
# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title 
plt.title('Two or more lines with different widths and colors with suitable legends ')
# Display the figure.
plt.plot(x1,y1, color='green', linewidth = 3,  label = 'line1-width-3')
plt.plot(x2,y2, color='skyblue', linewidth = 5,  label = 'line2-width-5')
# show a legend on the plot
plt.legend()
plt.show()
```
