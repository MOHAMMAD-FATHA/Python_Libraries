```python
"""
* @Author: Mohammad Fatha
* @Date: 2021-10-07 00:15
* @Last Modified by: Mohammad Fatha
* @Last Modified time: 2021-10-07 00:15
* @Title: : Python program to plot two or more lines and set the line markers.
"""

import matplotlib.pyplot as plt
# line 1 points
x1 = [10,20,30]
y1 = [20,40,10]
# line 2 points
x2 = [10,20,30]
y2 = [40,10,30]
# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Plot lines and/or markers to the Axes.
plt.plot(x1,y1, color='blue', linewidth = 1,  label = 'line1-dotted', marker = 'o')
plt.plot(x2,y2, color='red', linewidth = 1,  label = 'line2-dashed', marker = 'H') 
# Set a title 
plt.title("Plot with two or more lines with different styles")
# show a legend on the plot
plt.legend()
# function to show the plot
plt.show()

#Size pf the graph
```
