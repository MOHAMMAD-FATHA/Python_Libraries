```python
"""
* @Author: Mohammad Fatha
* @Date: 2021-10-09 20:18
* @Last Modified by: Mohammad Fatha
* @Last Modified time: 2021-10-09 20:18
* @Title: : Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv
"""

import seaborn as sns

df_train = sns.load_dataset('titanic')

sns.pointplot(x='sex',y='survived',data=df_train)
```
